```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
import matplotlib as mpl
import matplotlib.ticker as ticker
import matplotlib as mpl

mpl.rcParams["text.usetex"] = True

Nch_result = np.loadtxt('./Jeongseok_Wong/b_Nch_Nk_13TeV.csv', delimiter=',', usecols=[1])
Nk_result = np.loadtxt('./Jeongseok_Wong/b_Nch_Nk_13TeV.csv', delimiter=',', usecols=[2])

fig = plt.figure()
ax = plt.axes()
fig.set_size_inches(35, 16.534, forward=True)

plt.plot(Nch_result, np.nan_to_num(Nk_result), linewidth=7)

ax.xaxis.set_minor_locator(ticker.MultipleLocator(10))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.1))

ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: '${:g}$'.format(y)))
plt.tick_params(axis='both',which='major',direction='in',width=2,length=30,labelsize=45)
plt.tick_params(axis='both',which='minor',direction='in',width=2,length=15,labelsize=45)
plt.grid(color='silver',linestyle=':',linewidth=5)
plt.xlabel(r'$ N_{ch} $',size=70)
plt.ylabel(r'$\langle N_k \rangle$',size=70)
plt.tight_layout()

fig.savefig('./Jeongseok_Wong/Nch_Nk_pp13TeV.png')
```
